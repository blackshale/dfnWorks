## read data file csv

```python
import numpy as np
import matplotlib.pyplot as plt
import mplstereonet

data = np.genfromtxt('/home/bjung/github/dfnWorks/hlwd-2021/data/Az_dip_KSRS_csv.csv',dtype=None,delimiter=',',
                   skip_header=1,usecols=(0,1,2))
        
print(data)
print(type(data))
```

```python
depth_m = data[:,0]
strikes = (data[:,1])
dips = (data[:,2])

print(strike[0])
```

## plot poles on stereonet

```python
[fig,ax] = mplstereonet.subplots(figsize=(9.8,6),projection='equal_area')

cax = ax.density_contourf(strikes, dips, measurement='poles')

ax.pole(strikes, dips,markersize=2)
ax.grid(True)
fig.colorbar(cax)

plt.show()
```

```python
[fig,ax] = mplstereonet.subplots(figsize=(9.8,6),projection='stereonet')

cax = ax.density_contourf(strikes, dips, measurement='poles')

ax.pole(strikes, dips,markersize=2)
ax.grid(True)
fig.colorbar(cax)

plt.show()
```
